```python
sen = "Âº±Â∞èÁöÑÊàë‰πüÊúâÂ§ßÊ¢¶ÊÉ≥!"
```

## Step1 Âä†ËΩΩ‰∏é‰øùÂ≠ò

```python
# ‰ªéHuggingFaceÂä†ËΩΩÔºåËæìÂÖ•Ê®°ÂûãÂêçÁß∞ÔºåÂç≥ÂèØÂä†ËΩΩÂØπÂ∫îÁöÑÂàÜËØçÂô®
tokenizer = AutoTokenizer.from_pretrained("uer/roberta-base-finetuned-dianping-chinese")
tokenizer
```

```python
# ‰ªéÊú¨Âú∞Âä†ËΩΩtokenizer
tokenizer = AutoTokenizer.from_pretrained("./roberta_tokenizer/")
tokenizer
```

## Step2 Âè•Â≠êÂàÜËØç

```python
tokens = tokenizer.tokenize(sen)
tokens
```

## Step3 Êü•ÁúãËØçÂÖ∏

```python
tokenizer.vocab
```

```python
tokenizer.vocab_size
```

## Step4 Á¥¢ÂºïËΩ¨Êç¢

```python
# Â∞ÜËØçÂ∫èÂàóËΩ¨Êç¢‰∏∫idÂ∫èÂàó
ids = tokenizer.convert_tokens_to_ids(tokens)
ids, len(ids)
```

```python
# Â∞ÜidÂ∫èÂàóËΩ¨Êç¢‰∏∫tokenÂ∫èÂàó
tokens = tokenizer.convert_ids_to_tokens(ids)
tokens
```

```python
# Â∞ÜtokenÂ∫èÂàóËΩ¨Êç¢‰∏∫string
str_sen = tokenizer.convert_tokens_to_string(tokens)
str_sen
```

###  Êõ¥‰æøÊç∑ÁöÑÂÆûÁé∞ÊñπÂºè

```python
# Â∞ÜÂ≠óÁ¨¶‰∏≤ËΩ¨Êç¢‰∏∫idÂ∫èÂàóÔºåÂèàÁß∞‰πã‰∏∫ÁºñÁ†Å
ids = tokenizer.encode(sen, add_special_tokens=True)
ids, len(ids)
```

```python
# Â∞ÜidÂ∫èÂàóËΩ¨Êç¢‰∏∫Â≠óÁ¨¶‰∏≤ÔºåÂèàÁß∞‰πã‰∏∫Ëß£Á†Å
str_sen = tokenizer.decode(ids, skip_special_tokens=False)
str_sen
```

# Fast / Slow Tokenizer

```python
sen = "Âº±Â∞èÁöÑÊàë‰πüÊúâÂ§ßDreaming!"
```

```python
slow_tokenizer = AutoTokenizer.from_pretrained("uer/roberta-base-finetuned-dianping-chinese", use_fast=False)
slow_tokenizer
```

```python
%%time
# ÂçïÊù°Âæ™ÁéØÂ§ÑÁêÜ
for i in range(10000):
    fast_tokenizer(sen)
```

```python
%%time
# ÂçïÊù°Âæ™ÁéØÂ§ÑÁêÜ
for i in range(10000):
    slow_tokenizer(sen)
```
