自定义工具

````python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field

# 定义参数 Schema
class AddArgs(BaseModel):
    a: float = Field(..., description="第一个数")
    b: float = Field(..., description="第二个数")

# 实际实现函数
def add_impl(a: float, b: float) -> str:
    return str(a + b)

# 构建工具
add_tool = StructuredTool.from_function(
    func=add_impl,
    name="add_tool",
    description="计算两个数的加法，并返回字符串结果。",
    args_schema=AddArgs,
)
````

🔹 关键点

MessagesPlaceholder 用来在 prompt 中插入对话历史。

RunnableWithMessageHistory 接收一个 chain，和一个 get_history 回调，用于维护会话上下文。

调用时要传 config={"configurable": {"session_id": "xxx"}}，用来区分不同用户/会话。

````python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain.memory import ChatMessageHistory

# 1. 定义 Prompt
prompt = ChatPromptTemplate.from_messages(
    [
      ("system", "你是一个友好的 AI 助手。"),
      MessagesPlaceholder(variable_name="history"),
      ("human", "{input}"),
    ]
)

prompt
````

````python
# 2. 把 prompt 和 llm 串起来
chain = prompt | gemini_model
chain
````

````python
# 3. 包装成带记忆的 runnable
store = {}  # 用来保存不同会话的历史记录

def get_history(session_id: str):
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

conversation = RunnableWithMessageHistory(
    chain,
    get_history,
    input_messages_key="input",
    history_messages_key="history",
)

conversation
````

````python
# 4. 调用
resp1 = conversation.invoke({"input": "你好啊！"}, config={"configurable": {"session_id": "user1"}})
print(resp1)
````

````python
resp2 = conversation.invoke({"input": "能记住我刚刚说的吗？"}, config={"configurable": {"session_id": "user1"}})
print(resp2)
````

### 2.2 自定义LLM模块



````python
from langchain_core.language_models.llms import LLM
from typing import List, Optional, Any
import json

class MyDummyLLM(LLM):
    """一个简单的 LLM：把输入 prompt 原样包进 JSON 返回"""

    def _call(self, prompt, stop, **kwargs: Any):
        result = {"echo": prompt}
        return json.dumps(result, ensure_ascii=False)  # 确保中文也正常

    @property
    def _identifying_params(self):
        return {}

    @property
    def _llm_type(self):
        return "my_dummy_llm"
````

### 3.2 向量数据库

````python
!pip install -q faiss-cpu
````

LangChain 在设计时就对\*\*主流向量数据库（Vector Store）\*\*做了统一封装，帮你隐藏了底层的 **编码/解码** 细节（也就是「文本 → 向量」「向量 → 文档」的过程）。

---

🔹 LangChain 封装的关键点

1. **统一接口**
   所有向量数据库（FAISS、Chroma、Pinecone、Weaviate、Milvus…）在 LangChain 里都遵循同一个接口：

   * `from_documents()`：输入 `Document` 列表（会自动调用 embedding 模型编码成向量并存入 DB）。
   * `similarity_search(query)`：输入查询字符串（会自动调用 embedding 模型把 query 编码成向量，再去 DB 查）。
   * `add_documents()` / `add_texts()`：新增文档时自动编码。
   * `save_local()` / `load_local()`：存取本地索引。

   👉 你不用自己去 `model.encode()` 再存向量，也不用手动写反查逻辑。

---

2. **编码过程被隐藏**
   举个例子，用 FAISS：

   ```python
   from langchain_community.vectorstores import FAISS
   from langchain_community.embeddings import HuggingFaceEmbeddings

   # 初始化 embedding 模型
   embeddings = HuggingFaceEmbeddings(model_name="BAAI/bge-m3")

   # 直接从 Document 构建向量库
   db = FAISS.from_documents(docs, embeddings)

   # 相似度检索
   results = db.similarity_search("苹果授粉", k=3)
   for r in results:
       print(r.page_content)
   ```

   * **内部流程**：

     * `from_documents()` → 自动调用 `embeddings.embed_documents()` 编码文本成向量。
     * 存储时同时保存了 **原文（page\_content + metadata）**。
     * `similarity_search(query)` → 自动调用 `embeddings.embed_query(query)` 得到 query 向量，再去做相似度计算。
     * 返回的还是 `Document` 对象，你直接就能看到原始文本。

---

3. **解码过程也被隐藏**

   * 传统做法：拿到相似向量 ID → 再手动查数据库 → 拼接原文。
   * LangChain 封装后：`similarity_search()` 已经帮你把「ID → 原始文本」这一步做好了。
   * 所以你拿到的是 **包含 `page_content` 和 `metadata` 的 `Document`**，不用再写额外的“反查”逻辑。

---

🔹 图解流程

```
query (文本) ──▶ Embedding.encode() ──▶ 向量查询 ──▶ 最近邻向量 ──▶ Document（含原文）
     ↑                                                                   │
     └────────────────── LangChain 自动封装 ──────────────────────────────┘
```

---

🔹 优点

* **接口统一**：换数据库只改一行 import，不改业务逻辑。
* **开发效率高**：不用操心底层向量存储格式。
* **RAG 场景友好**：直接拿到原始文档对象，能马上送进 Prompt。

---

✅ 总结：
LangChain 的向量数据库封装，核心就是 **把「编码 → 存储」和「向量 → 文本解码」打包成统一接口**，你只需要关心 **输入文本** 和 **返回文档**。

### 4.2 LLMRouterChain 多链

````python
# pip install -q langchain langchain-community

from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.chains.router import MultiPromptChain
from langchain.chains.router.llm_router import LLMRouterChain, RouterOutputParser

# ==== 你的子链 ====
math_prompt = PromptTemplate.from_template("请解答下面的数学题：{input}")
story_prompt = PromptTemplate.from_template("请写一个简短的故事，主题是：{input}")
trans_prompt = PromptTemplate.from_template("请把下面的句子翻译成英文：{input}")

math_chain = LLMChain(llm=gemini_model, prompt=math_prompt)
story_chain = LLMChain(llm=gemini_model, prompt=story_prompt)
translate_chain = LLMChain(llm=gemini_model, prompt=trans_prompt)

destination_chains = {
    "math": math_chain,
    "story": story_chain,
    "translate": translate_chain,
}

# ==== 修复方案1：构建路由提示模板 ====
from langchain.chains.router.multi_prompt_prompt import MULTI_PROMPT_ROUTER_TEMPLATE

destination_infos = [
    {
        "name": "math",
        "description": "擅长处理数学计算问题"
    },
    {
        "name": "story",
        "description": "擅长写故事"
    },
    {
        "name": "translate",
        "description": "擅长翻译任务"
    }
]

# 构建destinations字符串
destinations_str = "\n".join([f"{info['name']}: {info['description']}" for info in destination_infos])

# 创建路由提示模板
router_template = MULTI_PROMPT_ROUTER_TEMPLATE.format(
    destinations=destinations_str
)

router_prompt = PromptTemplate(
    template=router_template,
    input_variables=["input"],
    output_parser=RouterOutputParser(),
)

# 创建路由链
router_chain = LLMRouterChain.from_llm(
    llm=gemini_model,
    prompt=router_prompt
)

default_chain = story_chain

multi_chain = MultiPromptChain(
    router_chain=router_chain,
    destination_chains=destination_chains,
    default_chain=default_chain,
    verbose=True,
)

# ==== 测试 ====
print("数学：", multi_chain.invoke({"input": "123 + 456 等于多少？"}))
print("故事：", multi_chain.invoke({"input": "小狗和小猫的冒险"}))
print("翻译：", multi_chain.invoke({"input": "今天天气真好"}))
````

### 4.3 Sequential Chain多链

````python
# pip install -q langchain langchain-community

from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.chains import SequentialChain

# 假设你已经有了 gemini_model
# from langchain_google_genai import ChatGoogleGenerativeAI
# gemini_model = ChatGoogleGenerativeAI(model="gemini-pro")

# ==== 基础 Sequential Chain 示例 ====
print("=== 基础 Sequential Chain 示例：故事创作流程 ===")

# 第一个链：生成故事大纲
outline_template = """请为以下主题创建一个简短的故事大纲：
主题：{topic}

故事大纲："""

outline_prompt = PromptTemplate(
    input_variables=["topic"],
    template=outline_template
)

outline_chain = LLMChain(
    llm=gemini_model,
    prompt=outline_prompt,
    output_key="outline"  # 指定输出变量名
)

# 第二个链：根据大纲写故事
story_template = """基于以下大纲，写一个完整的短篇故事：
大纲：{outline}

故事："""

story_prompt = PromptTemplate(
    input_variables=["outline"],
    template=story_template
)

story_chain = LLMChain(
    llm=gemini_model,
    prompt=story_prompt,
    output_key="story"  # 指定输出变量名
)

# 第三个链：为故事写评价
review_template = """请对以下故事进行评价和总结：
故事：{story}

评价："""

review_prompt = PromptTemplate(
    input_variables=["story"],
    template=review_template
)

review_chain = LLMChain(
    llm=gemini_model,
    prompt=review_prompt,
    output_key="review"  # 指定输出变量名
)

# 创建 Sequential Chain
story_sequential_chain = SequentialChain(
    chains=[outline_chain, story_chain, review_chain],
    input_variables=["topic"],  # 整个链的输入变量
    output_variables=["outline", "story", "review"],  # 想要输出的变量
    verbose=True
)

# 测试基础链
result = story_sequential_chain.invoke({"topic": "时间旅行的冒险"})
print("大纲：", result["outline"])
print("\n故事：", result["story"])
print("\n评价：", result["review"])
````

用途: 计算统计信息（平均值、中位数、极值）
场景: 数据分析、报表生成

````python
# pip install -q langchain langchain-community

from langchain.chains import TransformChain, LLMChain
from langchain.chains.combine_documents.stuff import StuffDocumentsChain
from langchain.chains.combine_documents.map_reduce import MapReduceDocumentsChain
from langchain.chains.combine_documents.refine import RefineDocumentsChain
from langchain.prompts import PromptTemplate
from langchain.docstore.document import Document
import re
import json

# 假设你已经有了 gemini_model
# from langchain_google_genai import ChatGoogleGenerativeAI
# gemini_model = ChatGoogleGenerativeAI(model="gemini-pro")

# ==========================================
# TRANSFORMATION CHAIN 示例
# ==========================================

print("=== TRANSFORMATION CHAIN 示例 ===\n")

# === 示例1：文本预处理转换链 ===
print("1. 文本预处理转换链")

def clean_text_transform(inputs: dict) -> dict:
    """清理和预处理文本"""
    text = inputs["text"]

    # 移除多余空格
    text = re.sub(r'\s+', ' ', text)
    # 移除特殊字符
    text = re.sub(r'[^\w\s\u4e00-\u9fff]', '', text)
    # 转为小写（英文部分）
    text = text.lower()
    # 统计字数
    word_count = len(text.split())

    return {
        "cleaned_text": text.strip(),
        "word_count": word_count,
        "original_length": len(inputs["text"]),
        "cleaned_length": len(text)
    }

# 创建文本清理转换链
text_clean_chain = TransformChain(
    input_variables=["text"],
    output_variables=["cleaned_text", "word_count", "original_length", "cleaned_length"],
    transform=clean_text_transform
)

# 测试文本清理
dirty_text = "  这是一个   有很多!!!空格和@#$特殊符号的文本...   需要清理！！！  "
clean_result = text_clean_chain.invoke({"text": dirty_text})

print(f"原始文本: '{dirty_text}'")
print(f"清理后文本: '{clean_result['cleaned_text']}'")
print(f"原始长度: {clean_result['original_length']}, 清理后长度: {clean_result['cleaned_length']}")
print(f"词数: {clean_result['word_count']}\n")
````

### 4.4 Document Chain
1. Stuff Documents Chain

原理: 将所有文档直接拼接后处理
适用: 文档较少、内容不太长的情况
优点: 简单直接，保持上下文

2. Map-Reduce Documents Chain

原理: 先分别处理每个文档（Map），再合并结果（Reduce）
适用: 文档较多、需要并行处理的情况
优点: 可扩展性好，适合大量文档

3. Refine Documents Chain

原理: 逐步精炼，用新文档改进之前的结果
适用: 需要迭代改进结果的场景
优点: 结果质量高，逐步优化

4. 自定义文档处理

用途: 提取元数据、统计分析、关键词识别
场景: 文档分析、内容管理

#### 4.4.1 Stuff Documents Chain

````python
print("=== DOCUMENT CHAIN 示例 ===\n")

# 创建示例文档
documents = [
    Document(
        page_content="人工智能（AI）是计算机科学的一个分支，旨在创建能够执行通常需要人类智能的任务的系统。AI技术包括机器学习、深度学习、自然语言处理等。",
        metadata={"source": "doc1", "topic": "AI基础"}
    ),
    Document(
        page_content="机器学习是AI的一个子集，它使计算机能够在没有明确编程的情况下学习和改进。常见的机器学习算法包括线性回归、决策树、神经网络等。",
        metadata={"source": "doc2", "topic": "机器学习"}
    ),
    Document(
        page_content="深度学习是机器学习的一个分支，使用多层神经网络来学习数据的复杂模式。它在图像识别、语音识别和自然语言处理方面取得了突破性进展。",
        metadata={"source": "doc3", "topic": "深度学习"}
    ),
    Document(
        page_content="自然语言处理（NLP）是AI的一个领域，专注于计算机与人类语言之间的交互。NLP技术用于机器翻译、文本分析、聊天机器人等应用。",
        metadata={"source": "doc4", "topic": "NLP"}
    )
]

# === 示例1：Stuff Documents Chain ===
print("1. Stuff Documents Chain (直接拼接)")

# 创建摘要提示模板
stuff_prompt = PromptTemplate(
    input_variables=["text"],
    template="""
请对以下文档内容进行总结，提取主要观点：

{text}

总结：
"""
)

# 创建LLM链
stuff_llm_chain = LLMChain(llm=gemini_model, prompt=stuff_prompt)

# 创建Stuff Documents Chain
stuff_chain = StuffDocumentsChain(
    llm_chain=stuff_llm_chain,
    document_variable_name="text",  # 在提示模板中使用的变量名
    verbose=True
)

# 运行Stuff Chain
print("Stuff Chain 结果:")
stuff_result = stuff_chain.invoke({"input_documents": documents})
print(stuff_result["output_text"])
print()
````

#### 4.4.2 Map-Reduce Chain

````python
print("2. Map-Reduce Documents Chain (分别处理后合并)")

# Map阶段：为每个文档创建摘要
map_template = """
总结以下文档的主要内容：
{text}

摘要：
"""
map_prompt = PromptTemplate(input_variables=["text"], template=map_template)
map_chain = LLMChain(llm=gemini_model, prompt=map_prompt)

# Reduce阶段：合并所有摘要
reduce_template = """
以下是多个文档的摘要，请将它们合并成一个连贯的总体摘要：

{text}

总体摘要：
"""
reduce_prompt = PromptTemplate(input_variables=["text"], template=reduce_template)
reduce_chain = LLMChain(llm=gemini_model, prompt=reduce_prompt)

# 创建Map-Reduce Documents Chain
map_reduce_chain = MapReduceDocumentsChain(
    llm_chain=map_chain,
    reduce_documents_chain=StuffDocumentsChain(
        llm_chain=reduce_chain,
        document_variable_name="text"
    ),
    document_variable_name="text",
    verbose=True
)

# 运行Map-Reduce Chain
print("Map-Reduce Chain 结果:")
map_reduce_result = map_reduce_chain.invoke({"input_documents": documents})
print(map_reduce_result["output_text"])
print()
````

#### 4.4.3 Refine Documents Chain

````python
print("3. Refine Documents Chain (逐步精炼)")

# 初始提示模板
initial_template = """
基于以下文档内容写一个摘要：
{text}

摘要：
"""
initial_prompt = PromptTemplate(input_variables=["text"], template=initial_template)

# 精炼提示模板
refine_template = """
你的任务是完善现有的摘要。
我们已经提供了一个到某个点的摘要：{existing_answer}

我们有机会通过下面的更多上下文来完善现有摘要：
{text}

基于新的上下文，完善原始摘要。如果上下文没有用处，返回原始摘要。

完善后的摘要：
"""
refine_prompt = PromptTemplate(
    input_variables=["existing_answer", "text"],
    template=refine_template
)

# 创建Refine Documents Chain
refine_chain = RefineDocumentsChain(
    initial_llm_chain=LLMChain(llm=gemini_model, prompt=initial_prompt),
    refine_llm_chain=LLMChain(llm=gemini_model, prompt=refine_prompt),
    document_variable_name="text",
    initial_response_name="existing_answer",
    verbose=True
)

# 运行Refine Chain
print("Refine Chain 结果:")
refine_result = refine_chain.invoke({"input_documents": documents})
print(refine_result["output_text"])
````

#### 4.4.4 Document Chain

````python
print("4. 自定义文档处理")

def custom_document_processor(docs):
    """自定义文档处理函数"""
    results = []

    for doc in docs:
        # 提取元数据
        topic = doc.metadata.get("topic", "未知")
        source = doc.metadata.get("source", "未知")

        # 统计内容
        content = doc.page_content
        char_count = len(content)
        word_count = len(content.split())

        # 查找关键词
        keywords = []
        key_terms = ["人工智能", "机器学习", "深度学习", "神经网络", "算法"]
        for term in key_terms:
            if term in content:
                keywords.append(term)

        results.append({
            "topic": topic,
            "source": source,
            "char_count": char_count,
            "word_count": word_count,
            "keywords": keywords,
            "content_preview": content[:50] + "..." if len(content) > 50 else content
        })

    return results

# 创建自定义转换链处理文档
def process_documents_transform(inputs: dict) -> dict:
    documents = inputs["documents"]
    processed = custom_document_processor(documents)

    # 生成统计报告
    total_docs = len(processed)
    total_chars = sum(doc["char_count"] for doc in processed)
    total_words = sum(doc["word_count"] for doc in processed)
    all_keywords = []
    for doc in processed:
        all_keywords.extend(doc["keywords"])
    unique_keywords = list(set(all_keywords))

    return {
        "processed_documents": processed,
        "total_documents": total_docs,
        "total_characters": total_chars,
        "total_words": total_words,
        "unique_keywords": unique_keywords,
        "report": f"处理了{total_docs}个文档，共{total_words}个词，{total_chars}个字符，发现{len(unique_keywords)}个唯一关键词。"
    }

# 创建自定义文档处理链
custom_doc_chain = TransformChain(
    input_variables=["documents"],
    output_variables=["processed_documents", "total_documents", "total_characters",
                     "total_words", "unique_keywords", "report"],
    transform=process_documents_transform
)

# 运行自定义处理
print("自定义文档处理结果:")
custom_result = custom_doc_chain.invoke({"documents": documents})
print(custom_result["report"])
print(f"发现的关键词: {custom_result['unique_keywords']}")

# 显示每个文档的详细信息
for i, doc_info in enumerate(custom_result["processed_documents"]):
    print(f"文档{i+1} - 主题: {doc_info['topic']}, 词数: {doc_info['word_count']}, 关键词: {doc_info['keywords']}")
````

## 05 记忆模块memory

| Memory 类型                        | 特点              | 适用场景          |
| -------------------------------- | --------------- | ------------- |
| `ConversationBufferMemory`       | 保存全部对话历史        | 短对话，调试方便      |
| `ConversationBufferWindowMemory` | 只保留最近 N 轮       | 长对话但只需关注近邻    |

### 5.1 ConversationBufferMemory

````python
from langchain.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.chains import LLMChain
from langchain.memory import ConversationBufferMemory


# 1. 定义 Prompt（带上历史 MessagesPlaceholder）
prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个友好的助手。"),
    MessagesPlaceholder(variable_name="history"),  # 用来插入记忆
    ("human", "{input}"),
])

# 2. 定义 Memory（保存对话历史）
memory = ConversationBufferMemory(return_messages=True, memory_key="history")

# 3. 创建 LLMChain
chain = LLMChain(llm=gemini_model, prompt=prompt, memory=memory, verbose=True)

# 4. 连续对话
print(chain.run("你好，我叫小明"))   # 第一次问候
print(chain.run("我叫什么名字？"))   # LLM 会从 memory 找到 "小明"
````

### 5.2 RunnableWithMessageHistory

````python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables.history import RunnableWithMessageHistory
from langchain.memory import ChatMessageHistory

# Prompt
prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个友好的助手。"),
    MessagesPlaceholder(variable_name="history"),
    ("human", "{input}")
])

chain = prompt | gemini_model

# 定义一个会话存储
store = {}
def get_history(session_id: str):
    if session_id not in store:
        store[session_id] = ChatMessageHistory()
    return store[session_id]

conversation = RunnableWithMessageHistory(
    chain,
    get_history,
    input_messages_key="input",
    history_messages_key="history"
)

# 测试对话
print(conversation.invoke({"input": "你好，我叫小明"}, config={"configurable": {"session_id": "user1"}}))
print(conversation.invoke({"input": "我叫什么名字？"}, config={"configurable": {"session_id": "user1"}}))
````

## 06 代理人 Agent

Agent拓展LLM的边界，从基础设施到应用。

| AgentType                                         | 决策范式            | 是否适合多轮对话       | 工具/函数要求                        | 是否易做结构化输出       | 优点              | 局限                         | 典型场景                |
| ------------------------------------------------- | --------------- | -------------- | ------------------------------ | --------------- | --------------- | -------------------------- | ------------------- |
| `ZERO_SHOT_REACT_DESCRIPTION`                     | ReAct（思考→行动→观察） | 一般（需手动加记忆）     | 需要给出**工具的自然语言描述**              | 一般              | 经典稳定，易理解与调试     | 长对话易“遗忘”；输出不够结构化           | 一次性问答、调工具查算         |
| `CONVERSATIONAL_REACT_DESCRIPTION`                | ReAct + 对话历史    | **是**（内置对话上下文） | 工具自然语言描述                       | 一般              | 多轮对话体验更好        | 仍然不强约束输出结构                 | 助手式工作流、连续追问         |
| `REACT_DOCSTORE`                                  | ReAct + 文档库     | 一般             | 需要 Docstore 工具（如 Wikipedia/自建） | 一般              | 对“查→读→答”路径清晰    | 依赖特定 docstore 工具           | 基于文档的问答             |
| `SELF_ASK_WITH_SEARCH`                            | 自提子问题 + 搜索      | 一般             | 搜索类工具                          | 一般              | 对开放域检索效果佳       | 更依赖搜索质量                    | 事实性问答、百科类问题         |
| `STRUCTURED_CHAT_ZERO_SHOT_REACT_DESCRIPTION`     | 结构化指令 + ReAct   | 一般（可配记忆）       | 工具自然语言描述                       | **较好**（更规范消息结构） | 更利于固定格式输出       | 复杂度略高                      | 需要清晰“思考/行动”信号与可解析输出 |
| `OPENAI_FUNCTIONS`（或 `OPENAI_TOOLS`）              | 函数/工具调用（函数签名）   | **是**（结合记忆最佳）  | **需要函数签名/参数模式**                | **最好**（严格 JSON） | 可靠的结构化工具调用与数据输出 | 依赖支持 functions/tools 的聊天模型 | 多工具编排、强结构化返回        |
| `CONVERSATIONAL_CHAT_ZERO_SHOT_REACT_DESCRIPTION` | 面向 Chat 的 ReAct | **是**          | 工具自然语言描述                       | 一般              | 兼顾对话与 ReAct     | 与 functions 相比结构化弱         | 聊天助手+调用少量工具         |

## 07 回调函数模块(callbacks)

LangChain 里 回调函数（Callbacks） 是一个非常重要的模块。它的作用是：在 LLM 推理 / Agent 执行 / 工具调用 的不同阶段，把事件“钩子”暴露出来，方便开发者记录日志、打印调试信息，甚至实现进度条、UI 展示等。

核心概念

CallbackHandler：定义如何处理事件的类，继承自 BaseCallbackHandler。

CallbackManager：统一管理一组 Handler。

触发时机：LLM 开始调用、LLM 返回结果、Agent 开始执行、Tool 开始执行、Tool 执行结束等。

````python
from langchain.callbacks.base import BaseCallbackHandler

# 自定义回调处理器
class MyCustomHandler(BaseCallbackHandler):
    def on_llm_start(self, serialized, prompts, **kwargs):
        print(">>> LLM 开始执行，Prompt 是：", prompts)

    def on_llm_end(self, response, **kwargs):
        print(">>> LLM 执行完成，返回内容：", response.generations[0][0].text)

    def on_tool_start(self, serialized, input_str, **kwargs):
        print(f">>> 工具 {serialized['name']} 开始执行，输入：{input_str}")

    def on_tool_end(self, output, **kwargs):
        print(">>> 工具执行完成，输出：", output)
````

````python
from langchain.agents import load_tools, initialize_agent, AgentType

# 工具
tools = load_tools(["llm-math"], llm=gemini_model)

# Agent
agent = initialize_agent(
    tools=tools,
    llm=gemini_model,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
    callbacks=[MyCustomHandler()]   # 绑定回调
)

agent.run("计算 25 的平方根。")
````
